```python
data=pd.read_csv('HCBG_Customer_dataset.csv',encoding='UTF8')
df=data.iloc[:,[5,9,24,37,43]]
data.head()
```

```python
numerical_df=data.iloc[:,list(range(49,59,1))]
```

```python
#convoptyss 지우기
numerical_df=numerical_df.drop("convoptyss",axis=1)
numerical_df.head()
```

```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt

x = []   # k 가 몇개인지 
y = []   # 응집도가 몇인지 

for k in range(1, 10):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(numerical_df)
    
    x.append(k)
    y.append(kmeans.inertia_)
    
plt.plot(x, y)
```

```python
import seaborn as sns
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=3)
model=kmeans.fit(numerical_df)
numerical_df['labels'] =model.predict(numerical_df)

sns.scatterplot(x='EmailSends', y='conv leads', hue='labels', data=numerical_df)
```

```python
numerical_df["labels"].value_counts()
```

```python
# 위의 clustering 결과를 3차원 시각화
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d') 

ax.set_xlabel('Emailsends')
ax.set_ylabel('conv leads')
ax.set_zlabel('formsubmits')

x = numerical_df['EmailSends']
y = numerical_df['conv leads']
z = numerical_df['formsubmits']
ax.scatter(x, y, z, c = numerical_df["labels"], s= 20, alpha=0.5, cmap='rainbow')
```
